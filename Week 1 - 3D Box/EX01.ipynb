## Exercise  1.12

```python
Q = box3d(16)
K = np.array([[1, 0, 0], [0, 1, 0], [0, 0, 1]])
# equal to the position of the camera facing the object
t = np.array([[0, 0, 4]]).T
R = K.copy()
Q = to_homogeneous(Q)
camera = to_inhomogeneous(project_points(K, R, t, Q))
print(len(camera), len(camera[0]))
plt.scatter(camera[0], camera[1])
plt.savefig('ex1_plot2.jpg', bbox_inches='tight', dpi=300)
plt.show()
```

## Exercise 1.13

```python
t = np.array([[0, 0, 4]]).T
theta = math.radians(30)
R = np.array([[math.cos(theta), 0, math.sin(theta)], [
             0, 1, 0], [-math.sin(theta), 0, math.cos(theta)]])
project_points(K, R, t, Q)
camera = to_inhomogeneous(project_points(K, R, t, Q))
plt.scatter(camera[0], camera[1])
plt.savefig('ex1_plot3.jpg', bbox_inches='tight', dpi=300)
plt.show()
```
